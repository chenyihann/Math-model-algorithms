```python
import numpy as np
import matplotlib.pyplot as plt

# 参数设置
b = 1.7/(2*np.pi)
theta = np.linspace(0, 32 * np.pi, 10000)  # θ从0到32π，增加点数使曲线更平滑

# 计算两条螺旋线的极径
r1 = b * theta                # 原曲线 r = bθ
r2 = b * (theta + np.pi)      # 对称曲线 r = b(θ + π)

# 创建极坐标子图
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='polar')

# 绘制两条螺旋线
ax.plot(theta, r1, label=r'$r = 1.7\theta$', color='blue', linewidth=1)
ax.plot(theta, r2, label=r'$r = 1.7(\theta + \pi)$', color='red', linewidth=1)

# 设置标题和图例
ax.set_title('', pad=20)
ax.legend(loc='upper right')

# 显示图形
plt.show()
```
